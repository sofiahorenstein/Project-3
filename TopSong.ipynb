```python
import pandas as pd
```

```python
spotifyDF = pd.read_csv("spotify1990.csv") 
billboardDF = pd.read_csv("billboard1990.csv") 
```

```python
spotifyDF.head()
```

```python
billboardDF.head()
```

```python
t=0
onBillboard = []
while t<len(spotifyDF['track']):
    song = spotifyDF['track'][t]
    s=0
    while s<len(billboardDF['song']):
        if song == billboardDF['song'][s]:
            onBillboard.append(song)
            break
        else:
            s=s+1
    t=t+1
```

```python
onBillboard
```

```python
uniqueBB = []
[uniqueBB.append(x) for x in onBillboard if x not in uniqueBB] 
uniqueBB
```

```python
len(uniqueBB)
```

```python
t=0
topSong = []
while t<len(spotifyDF['track']):
    song = spotifyDF['track'][t]
    s=0
    i=0
    while s<len(uniqueBB):
        if song == uniqueBB[s]:
            i=i+1
        s=s+1
    topSong.append(i)
    t=t+1
topSong
```

```python
topSong.count(1)
```

```python
spotifyDF['top_song']=topSong
```
