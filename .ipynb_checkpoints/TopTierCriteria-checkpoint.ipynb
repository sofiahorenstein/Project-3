```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Datasets

```python
data = pd.read_csv("spotifyTOP.csv") 
data.head()
```

```python
# Drop irrelevant columns
# 11,000 songs
data2 = data.drop(['Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 0.1.1', 'id', 'duration_ms', 'key', 'mode', 'time_signature'], axis=1)
data2.head()
```

```python
# Fetch songs that made the 'Top'
# 'top_song' column includes: 0 = 'no', 1 = 'yes'
data3 = data2.loc[data2['top_song'] == 1]

# Reset index
data4 = data3.reset_index(drop=True)

top = data4.drop(['top_song'], axis=1)

top.head()
```

```python
# Fetch songs that did not make the 'Top'
data5 = data2.loc[data2['top_song'] == 0]

# Reset index
data6 = data5.reset_index(drop=True)

not_top = data6.drop(['top_song'], axis=1)

not_top.head()
```

### Analysis

```python
top_analysis = top.describe()
top_analysis
```
