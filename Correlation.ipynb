```python
import pandas as pd
```

```python
spotify_top=pd.read_csv("spotifyTOP.csv")
spotify_top.head()
```

```python
spotify_top_2=spotify_top[spotify_top['top_song']==1]
spotify_top_2.head()
```

```python
billboard_join=pd.read_csv("billboard_join.csv")
billboard_join.head()
```

```python
billboard_join_2=billboard_join.rename(columns={"song":"track"})
billboard_join_2.head()
```

```python
spotify_billboard=spotify_top_2.merge(billboard_join_2, on=["track", "artist"], how="left")
spotify_billboard.head()
```

```python
spotify_billboard_2=spotify_billboard.drop_duplicates(subset=["id"])
spotify_billboard_2.head()
```

```python
spotify_billboard_3=spotify_billboard_2.dropna(subset=["year"])
spotify_billboard_3.head()
```

```python
spotify_billboard_4 = spotify_billboard_3.loc[:, ~spotify_billboard_3.columns.str.contains('^Unnamed')]
spotify_billboard_4.head()
```

```python
spotify_billboard_drop=spotify_billboard_4.drop(columns=['year', 'track', 'artist', 'id', 'top_song', 'mode', 'duration_ms', 'tempo', 'popularity', 'loudness', 'time_signature', 'key'])
spotify_billboard_drop.head()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
a4_dims = (10, 4)
fig, ax = plt.subplots(figsize=a4_dims)
sns.catplot(ax=ax, data=spotify_billboard_drop, kind="box")
plt.tight_layout()
ax.set_title('Songs on Billboard Top 100', fontsize=18)
fig.savefig("boxplot1.jpeg")
```
